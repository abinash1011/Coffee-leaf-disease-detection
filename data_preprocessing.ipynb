# make test data

```python
import os
import random
import shutil

# Define the paths to your training and test data folders
path_to_training_data = 'dataset128'
path_to_test_data = 'small_dataset'
num_test_images = 5

# Create the test data folder structure if it doesn't exist
if not os.path.exists(path_to_test_data):
    os.makedirs(path_to_test_data)

# Define the ratio for test data (e.g., 70% training, 30% test)
# test_data_ratio = 0.3

# Iterate through each class folder in the training data
for class_folder in os.listdir(path_to_training_data):
    class_path = os.path.join(path_to_training_data, class_folder)

    # Create a corresponding class folder in the test data
    test_class_path = os.path.join(path_to_test_data, class_folder)
    if not os.path.exists(test_class_path):
        os.makedirs(test_class_path)

    # List all image files in the class folder
    image_files = os.listdir(class_path)

    # Calculate the number of images to move to the test data
    # num_test_images = int(len(image_files) * test_data_ratio)

    # Randomly select and move images to the test data folder
    test_images = random.sample(image_files, num_test_images)
    for test_image in test_images:
        src_path = os.path.join(class_path, test_image)
        dest_path = os.path.join(test_class_path, test_image)
        
        shutil.move(src_path, dest_path)
```

# Count datapoints

```python
image_directory = "dataset224/Phoma"
```

```python
image_paths = []
image_widths = []
image_heights = []
for filename in os.listdir(image_directory):
    if filename.endswith((".jpg", ".png", ".jpeg")):  # You can add more file extensions if needed
        # Construct the full path to the image
        image_path = os.path.join(image_directory, filename)
        
        # Load the image using OpenCV
        img = cv2.imread(image_path)
        
        if img is not None:  # Ensure the image was successfully loaded
            height, width, _ = img.shape  # Get image dimensions
            
            # Append the image path and dimensions to the lists
            image_paths.append(image_path)
            image_widths.append(width)
            image_heights.append(height)
data = {
    "Image Path": image_paths,
    "Width": image_widths,
    "Height": image_heights
}

df = pd.DataFrame(data)
print(df.Height.value_counts())
df.Width.value_counts()
```

# resize from dir

```python
from PIL import Image
import os

# Define the input directory containing your images
input_directory = "datasets/Phoma"

# Define the output directory where resized images will be saved
output_directory = "dataset224/Phoma"

# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

# Define the target size (128x128 pixels)
target_size = (224, 224)

# Loop through the images in the input directory
for filename in os.listdir(input_directory):
    # Construct the full path to the image
    image_path = os.path.join(input_directory, filename)

    try:
        # Open the image using PIL
        img = Image.open(image_path)

        # Check the current size of the image
        current_size = img.size

        # Check if the image size matches the target size
        if current_size != target_size:
            # Calculate the new size with the same aspect ratio
            img = img.resize(target_size, Image.ANTIALIAS)

        # Save the (resized or original) image to the output directory
        output_path = os.path.join(output_directory, filename)
        img.save(output_path)
        print(f"Saved {filename} to {output_path}")

    except Exception as e:
        print(f"Error processing {filename}: {e}")
```
